```python
##代码结构
#1.访问功能--首先打开我们需要的第一个页面
#2.构造函数，可以提取任意当前页面的所有内容
#3.开发翻页功能，不断调用步骤2中的函数，然后将数据整合到一起
#4.将步骤3中的功能也打包成函数，然后可以替换任意城市，替换不同的工作岗位
```

```python
#brower已经正确了，直接进行提取
def get_lagou_one_page():
    #新建一个空的DataFrame
    df_one=pd.DataFrame()
    title=[i.text for i in driver.find_elements_by_tag_name('h3')]
    area=[i.find_element_by_tag_name('em').text for i in driver.find_elements_by_class_name('add')]
    time=[i.text for i in driver.find_elements_by_class_name('format-time')]
    money=[i.text for i in driver.find_elements_by_class_name('money')]
    demand=[i.find_element_by_class_name('li_b_l').text for i in driver.find_elements_by_class_name('p_bot')]
    label=[i.text for i in driver.find_elements_by_xpath('//div[@class="list_item_bot"]/div[@class="li_b_l"]')]
    welfare=[i.text for i in driver.find_elements_by_class_name('li_b_r')]
    industry=[i.text for i in driver.find_elements_by_class_name('industry')]
    company_name=[i.find_element_by_tag_name('a').text for i in driver.find_elements_by_class_name('company_name')]
    #将上面提取出来的数据保存进来
    df_one['职位名称']= title
    df_one['公司地址']= area
    df_one['发布时间']= time
    df_one['薪资待遇']= money
    df_one['岗位要求']= demand
    df_one['岗位标签']= label
    df_one['公司福利']= welfare
    df_one['所属行业']= industry
    df_one['公司名称']= company_name
    return df_one
```

```python
df_one=get_lagou_one_page()
df_one
```

```python
#建一个空的DataFrame用于合并
df=pd.DataFrame()
for i in range(30):
    df_one = get_lagou_one_page()
    #内容合并
    df=df.append(df_one,ignore_index=True)
    #翻页
    driver.find_element_by_xpath('//span[@action="next"]').click()
    time.sleep(6)
```

```python
#将上面的功能打包成函数
def get_lagou_data(job_name,city_num,page_num):
    #1.访问第一页，构造URL
    url="https://www.lagou.com/jobs/list_{}/p-city_{}?px=default#filterBox".format(job_name,city_num)
    driver.get(url)
    time.sleep(6)#手动点掉出现的广告
    df=pd.DataFrame()
    #开始循环
    for i in range(page_num):
        df_one=get_lagou_one_page()
        #内容合并
        df=df.append(df_one,ignore_index=True)
        #翻页
        driver.find_element_by_xpath('//span[@action="next"]').click()
        time.sleep(6)
    return df 
```
