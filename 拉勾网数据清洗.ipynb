```python
#根据薪资查找异常值
def bool_out_scope(x, times=3):
    x = pd.Series(x)
    lb, ub = x.mean() + pd.Series([-1,1]) * times * x.std()
    return (x> ub) | (x<lb)
```

```python
df[bool_out_scope(df.平均薪资)]
```

```python
df[bool_out_scope(df.平均薪资) & df.城市.isin(['beijin','shanghai','shenzhen'])]
```
